```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd



```

```python
desde="2010-01-01"
hasta="2016-12-31"

amazon=yf.download(tickers="AMZN", start=desde, end=hasta)


print(amazon.head())
print(type(amazon))
```

```python
cierre=amazon["Close"]
cierre
```

```python
fig=plt.figure(figsize=(14,5))
ax=fig.add_subplot()

ax.plot(amazon.index,amazon["Close"], color="tab:brown")
ax.set_title("Cotizacion de las acciones de Amazon", size="x-large",color="tab:blue")
```
