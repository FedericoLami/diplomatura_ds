```python
# 3)Buscar cuáles son las canciones con tempos (*bpm*, pulsaciones por minuto) más elevados.
df_bpm=df.sort_values(by="bpm", ascending=False).head(10)
df_bpm
```

```python
# 4)Crear un nuevo DataFrame con los datos de ritmo (bpm) y duración (dur) promedio para cada año. 
a=df.groupby(by='top year').mean("bpm")
df_bpm_dur=pd.DataFrame({'bpm':a["bpm"],'dur':a["dur"]})
df_bpm_dur
```

```python
#Creamos graficos con el dataframe anterior.

a.plot.line(y="bpm")
a.plot.line(y="dur")
```
