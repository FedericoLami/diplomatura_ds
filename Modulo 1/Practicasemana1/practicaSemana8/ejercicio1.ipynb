```python
# calculamos la media, desviacion estandar, valor minimo y valor maximo de la columna "altura"

alturas=estaciones["ALTURA"]
print(" Media: ",round(alturas.mean(),2),"\n","Desviacion estandar: ", round(alturas.std(),2),"\n", "Valor maximo: ", alturas.max(),"\n", "Valor minimo: ", alturas.min())
```
