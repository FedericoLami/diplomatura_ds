```python
import numpy as np
import pandas as pd
mi_lista = list('abcdefghijklmnopqrstuvwxyz')

mi_array = np.arange(26)

mi_diccionario = dict(zip(mi_lista,mi_array))
```

```python
# Creamos series de pandas con los datos en el codigo de arriba. Para eso, usamos el metdo pd.Series()

serLis=pd.Series(mi_lista)
serArr=pd.Series(mi_array)
serDic=pd.Series(mi_diccionario)
```

```python
#Convertimos serDic en un dataframe con su indice como otra columna de dataframe.

serDicDataframe=pd.DataFrame({"letra": serDic.index, "valor": serDic.values })
serDicDataframe
```

```python
# combinamos serLis y serArr para formar un dataframe

LisArrDf=pd.DataFrame({"serLis":serLis,"serArr":serArr})
LisArrDf
```

```python
#exportamos el dataframe LisArrDf a un archivo con nombre serDic.csv

LisArrDf.to_csv(r"C:\Users\mati_\Desktop\Diplomatura_en_ciencia_de_datos\serDic.csv", index=False)
```
