```python
#Creamos una nueva serie que contenga los datos de las personas de primera clase

primera=titanic[titanic["Clase"]==1]

#reseteamos el indice

primera=primera.reset_index()

#ahora le pedimos que almacene los nombres en la variable nombres
nombres=primera["Nombre"]

#ordenamos los nombres alfabeticamente
nombres=nombres.sort_values()
print("Los nombres de las personas en primera clase:\n", nombres)
```

```python
#Eliminamos los datos de las personas que no figura la edad

personas=titanic.dropna(subset="Edad")
personas
```

```python
#determinamos la edad media de las mujeres que viajaban en 1era clase

mujeres=personas[(personas["Genero"]=="female") & (personas["Clase"]==1)]
promedad=mujeres["Edad"].mean()
promedad
print("La edad promedia de las mujeres que viajaban en primera clase es: \n", round(promedad,2))
```

```python
#Creamos un histograma con los sobrevivientes. Para eso, primero vamos a armar un nuevo dataframe con los sobrevivientes.

sobrevivientes=titanic[titanic["Sobrevivencia"]==1]
sobrevivientes

#ahora creamos el histograma

sobrevivientes.Edad.plot.hist( bins=80, figsize= (15,7))
sobrevivientes['Edad'].plot.hist(bins=20)
```
